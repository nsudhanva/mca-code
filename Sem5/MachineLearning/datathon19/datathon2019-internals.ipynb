```python
from sklearn.preprocessing import LabelEncoder
df = df.apply(LabelEncoder().fit_transform)
df
```

```python
X = df.iloc[:, 0:9].values
y = df.iloc[:, 9].values
```

```python
print(X.shape)
print(y.shape)
print(df.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.308, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion = 'entropy', max_depth = 5, min_samples_split = 10)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
y_pred = model.predict(X_test)
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
y_pred.shape
```

```python
y_pred = pd.DataFrame(y_pred)
y_pred.replace(0, 'positive', inplace=True)
y_pred.replace(1, 'negative', inplace=True)

y_pred['Id'] = y_pred.index

y_pred.columns = ['class', 'Id']
# y_pred.reindex(columns=['Id', 'class'])
y_pred.head()
```

```python
y_pred = y_pred[['Id', 'class']]
```
